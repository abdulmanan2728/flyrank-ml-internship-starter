## 2. The question: decision, action, cost of a wrong call

**Decision:** which pages should an SEO analyst review this week, out of potentially
thousands.

**Who acts on it:** an SEO analyst doing the weekly content review.

**Output:** a ranked list of pages, each scored and reason-coded, feeding a weekly
review queue.

**Cost of a wrong call:** a false positive (flagging a page that's actually fine)
wastes analyst time on something that didn't need review — the mistake I'd rather bias
away from. A false negative (missing a genuinely declining page) means it keeps losing
visibility unnoticed for another review cycle — a real but recoverable cost, since it
just waits for next week's queue. Neither is catastrophic, which is why this is
decision-support, not an automated action.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

# Number 1: scale — how many content items, across how many clients
print(f"content items: {df.shape[0]}, unique clients: {df['client_id'].nunique()}")

# Number 2: base rate of decline — is this a common enough problem to matter?
declining_share = (df['trend_direction'] == 'down').mean()
print(f"share of pages currently trending down: {declining_share:.3f}")

# Number 3: how blunt a naive single-signal rule would be, vs. the real label —
# motivating why this needs more than an if-statement
naive_flagged = (df['ctr'] < df['ctr'].median()).mean()
print(f"naive rule (below-median CTR) flags: {naive_flagged:.3f} vs actual declining rate: {declining_share:.3f}")
```

## 4. Careful words: what I can and can't claim

**What I can claim:** observed, measured, directional results — e.g. "pages with these
characteristics are more likely to be trending down in this dataset," and
decision-support recommendations for which pages to prioritize reviewing.

**What I will never claim:** that this predicts or explains Google's ranking algorithm,
that a flagged page is *causally* declining because of any one feature, or that the
model's output should replace human judgment rather than prioritize it. All results are
directional and tied to FlyRank's anonymized measurement of search performance — not a
universal truth about search visibility.
